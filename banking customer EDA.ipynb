```python

bins = [0,100000,300000,float('inf')]
labels = ['Low','Med','High']
df['income_band'] =  pd.cut(df['estimated_income'], bins = bins, labels = labels)
```

```python
df['income_band'].value_counts().plot(kind = 'bar')
```

```python
# brid is type of bankng institution id
brid_df = {1 :'Retail',
        2:'Insittutional',
        3:'Private',
        4:'Commercial'}
df['brid'] = df['brid'].map(brid_df)
```

```python
# genderid
genderid_df=  {1:'Male',2:'Female'}
df['genderid'] = df['genderid'].map(genderid_df)
```

```python
# separating columns 
categorical_columns = ['brid', 'genderid', 'iaid','income_band',  'nationality',
                        'fee_structure','loyalty_classification',  'amount_of_credit_cards','properties_owned','risk_weighting'     ]
numerical_columns =  ['estimated_income', 'superannuation_savings','credit_card_balance','bank_loans','bank_deposits','checking_accounts','saving_accounts', 'foreign_currency_account', 'business_lending',]
```
